```python
# Importing Of Libraries
import pandas as pd
import math
import seaborn as sns
import numpy as np
from sklearn import linear_model
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
# Clearing Empty Values From Dataset
df_trainingSetDirty = pd.read_csv('LinearRegressionTrain.csv')
df_testSetDirty = pd.read_csv('LinearRegressionTest.csv')
trainingSet = df_trainingSetDirty.dropna()
testSet = df_testSetDirty.dropna()
print ("Rows before clean: ", df_trainingSetDirty.size, "\n")
print ("Rows after clean: ", trainingSet.size, "\n")
```

```python
# Making Data Ready For Scikit Learn as it takes 2D Matrix.
x_training_set = trainingSet.as_matrix(['x'])
y_training_set = trainingSet.as_matrix(['y'])

x_test_set = testSet.as_matrix(['x'])
y_test_set = testSet.as_matrix(['y'])
```

```python
# Observation in Data which includes Bias As Well As Variance which gives result about Positive Skew or Negative Skew
print ("Mean of X Training set: ", np.mean(x_training_set), "\n")
print ("Median of X Training set: ", np.median(x_training_set), "\n")
print ("Mean of Y Training set: ", np.mean(y_training_set), "\n")
print ("Median of Y Training set: ", np.median(y_training_set), "\n")
print ("Std Dev of X Training set: ", np.std(x_training_set), "\n")
print ("Std Dev of Y Training set: ", np.std(y_training_set), "\n")

# As there is not much difference between mean and median which shows Positive Skew which is good.
```

```python
# Making Linear Regression object as well as training of Data is Done
reg = linear_model.LinearRegression()
reg.fit(x_training_set,y_training_set)
```

```python
# Plotting of Training Data from as well as predicted data from model
plt.xlabel('X')
plt.ylabel('Y')
plt.legend('XY Graph')
plt.scatter(x_training_set,y_training_set,color='yellow',marker='x')
plt.plot(x_training_set,reg.predict(x_training_set),color='black')
plt.show()
```

```python
# R Square and Correlation Coefficient
print('R sq: ',reg.score(x_training_set,y_training_set))
print('correlation: ',math.sqrt(reg.score(x_training_set,y_training_set)))
```

```python
# So let's plot some of the data 
# - this gives some core routines to experiment with different parameters
plt.title('Relationship between X and Y')
plt.scatter(x_training_set, y_training_set,  color='black')
plt.show()

# Use subplot to have graphs side by side
plt.subplot(1, 2, 1)
plt.title('X training set')
plt.hist(x_training_set)

plt.subplot(1, 2, 2)
plt.title('Y training set')
plt.hist(y_training_set)
plt.show()

plt.subplot(1, 2, 1)
plt.title('X training set')
plt.boxplot(x_training_set)

plt.subplot(1, 2, 2)
plt.title('Y training set')
plt.boxplot(y_training_set)

plt.show()
sns.distplot(x_training_set)
```
